### Imports

```python
import requests
import os
from dotenv import load_dotenv
```

### Env variables

```python
load_dotenv()

nuvemshop_token = os.getenv("NUVEMSHOP_API_TOKEN")
vdl_user_id = os.getenv("VDL_USER_ID")
nuvemshop_base_url = os.getenv("NUVEMSHOP_BASE_URL")
```

### Função para pegar Pedidos da VDL

```python
def get_orders(nuvemshop_base_url, token, user_id):
    endpoint_url = f"{nuvemshop_base_url}/{user_id}/orders"
    headers = {
        "Authentication": f"bearer {token}",
        "User-Agent": "VDL Solutions 1.0 (leobelko@gmail.com)"
    }

    try:
        response = requests.get(endpoint_url, headers=headers)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        raise e

    return response
```

Testando:

```python
print(get_orders(nuvemshop_base_url, nuvemshop_token, vdl_user_id))
```
